# Tensor Broadcasting Playground

Get better at pytorch's broadcasting rules

```python
import torch
```

```python
traj = torch.rand(3,5,2)
thresh = torch.rand(5)
traj, thresh
```

```python
init = traj[:, 0, :]
init
```

```python
input_from_init = traj - init.unsqueeze(1)
norm_from_init = torch.norm(input_from_init, dim=-1)
norm_from_init
```
